```python
from google_play_scraper import app, reviews
import pandas as pd

# 앱 정보 가져오기
package_name = "com.velox.hkmc_tm1k"  # 패키지 이름
app_info = app(package_name)
print("앱 이름:", app_info['title'])

review_list = []

# 앱 리뷰 가져오기
for i in range(1,5) :
    reviews_result, _ = reviews(
        package_name,
        lang='ko',  # 한국어
        country='kr',  # 한국 사용자
        count=400,  # 가져올 리뷰 수
        filter_score_with=i  # 특정 평점 필터링 (None: 모든 평점)
    )
    for review in reviews_result:
        review_list.append({'rating' : review['score'], 'review' : review['content']})

reviews_result, _ = reviews(
    package_name,
    lang='ko',  # 한국어
    country='kr',  # 한국 사용자
    count=800,  # 가져올 리뷰 수
    filter_score_with=5  # 특정 평점 필터링 (None: 모든 평점)
)
for review in reviews_result:
    review_list.append({'rating' : review['score'], 'review' : review['content']})


review_data = pd.DataFrame(review_list)
```

```python
from wordcloud import WordCloud
from konlpy.tag import Okt

text = " ".join(review_data['review'].astype(str))
okt = Okt()

# 조사 제거 함수
def remove_josa(text):
    tokens = okt.pos(text)  # 형태소 분석 및 품사 태깅
    filtered_tokens = [word for word, tag in tokens if tag != 'Josa']  # 조사를 제외
    return " ".join(filtered_tokens)  # 다시 문장으로 합치기

# 조사 제거 적용
processed_text = remove_josa(text)

processed_text
```

```python
wordcloud = WordCloud(font_path='/Users/admin/Library/Fonts/NanumGothic.otf',max_words=200).generate(text)
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.title("Word Cloud")
plt.axis('off')
plt.show()
```
