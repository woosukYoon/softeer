```python
from multiprocessing import Pool
import time

def work_log(work):

    name, duration = work
    start = int(time.time())
    print(name, ': ', duration, 'waiting')
    time.sleep(duration)
    print(name, 'finished')

def main() :
    work = [('A', 5), ('B', 2), ('C', 1), ('D', 3)]
    pool = Pool(2)
    pool.map(work_log, work)
    pool.close()
    pool.join()

if __name__ == "__main__":
    main()
```
